```python
import torch 
import torch.nn as nn
```

```python
class customMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(10,32)
        self.layer2 = nn.Linear(32,16)
        self.layer3 = nn.Linear(16,2)
    
    def forward(self, x):
        x1 = torch.relu(self.layer1(x))
        x2 = torch.relu(self.layer2(x1))
        x3 = self.layer3(x2)
        print(x1.shape)
        print(x2.shape)
        print(x3.shape)
        return x
```

```python
model = customMLP()
x = torch.randn(32,10)
output = model(x)
```
