```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1.0/(1.0+torch.exp(-x))

def relu(x):
    return torch.maximum(torch.tensor(0.0),x)

def gelu(x):
    part1 = torch.sqrt(torch.tensor(2)/np.pi) * (torch.pow(x,3) * 0.044715 + x)
    return 0.5 * x * (1.0 + torch.tanh(part1))
```

```python
def main():
    x = torch.linspace(-5,5,100)

    sigmoid_y = sigmoid(x)
    relu_y = relu(x)
    gelu_y = gelu(x)

    plt.figure(figsize=(18,5))
    #convert back to numpy for plotting with matplot
    plt.subplot(1,3,1)
    plt.plot(x.numpy(),sigmoid_y.numpy(), color = 'b')
    plt.xlabel("Input x")
    plt.ylabel("Output activation y")
    plt.grid(True)
    plt.axhline(0, color="black")
    plt.title("Sigmoid")

    plt.subplot(1,3,2)
    plt.plot(x.numpy(),relu_y.numpy(), color = 'r')
    plt.xlabel("Input x")
    plt.ylabel("Output activation y")
    plt.grid(True)
    plt.axhline(0, color="black")
    plt.title("reLU")

    plt.subplot(1,3,3)
    plt.plot(x.numpy(),gelu_y.numpy(), color = 'g')
    plt.xlabel("Input x")
    plt.ylabel("Output activation y")
    plt.grid(True)
    plt.axhline(0, color="black")
    plt.title("GELU")
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
