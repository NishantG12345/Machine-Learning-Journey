```python
import torch
import math
```

```python
a = torch.tensor(2.0, requires_grad = True)
b = torch.tensor(3.0, requires_grad = True)
c = a * b
d = torch.sin(c)
#backprop
d.backward()
```

```python
def manual_backprop(a, b):
    dd_da = math.cos(a * b) * b
    dd_db = math.cos(a * b) * a 
    return dd_da, dd_db

```

```python
dd_da, dd_db = manual_backprop(a.item(),b.item())
print(f"manual derivative w respect to a{dd_da}")
print(f"manual derivative w respect to b{dd_db}")
print(f"pytorch derivative w respect to a{a.grad.item()}")
print(f"pytorch derivative w respect to b{b.grad.item()}")
```
